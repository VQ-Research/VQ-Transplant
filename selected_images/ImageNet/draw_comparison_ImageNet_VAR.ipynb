```python
import os
import json
from PIL import Image, ImageDraw, ImageFont
from pprint import pprint
```

```python
manul_picked = [
    "004339.png",
    "013772.png",
    "016954.png",
    "048662.png",
    "047779.png",
    "011531.png",
    "017008.png",
    "016995.png"
]
```

```python
original_images = []
var_images = []
mmd_var_images = []
wasserstein_var_images = []
online_var_images = []
vanilla_var_images = []
ema_var_images = []


# Base directory for images
base_dir = 'D:\ResearchProject\MLProjects\VQ-Transplant\\figures\ImageNet'


for key in manul_picked:
    # image path
    original_img_path = os.path.join(base_dir, 'Original_pick', key)
    var_img_path = os.path.join(base_dir, 'VAR', key)
    mmd_var_img_path = os.path.join(base_dir, 'MMD_VAR', key)
    wasserstein_var_img_path = os.path.join(base_dir, 'Wasserstein_VAR', key)
    online_var_img_path = os.path.join(base_dir, 'Online_VAR', key)
    ema_var_img_path = os.path.join(base_dir, 'EMA_VAR', key)
    vanilla_var_img_path = os.path.join(base_dir, 'Vanilla_VAR', key)

    # Load images
    original_img = Image.open(original_img_path)
    var_img = Image.open(var_img_path)
    mmd_var_img = Image.open(mmd_var_img_path)
    wasserstein_var_img = Image.open(wasserstein_var_img_path)
    online_var_img = Image.open(online_var_img_path)
    ema_var_img = Image.open(ema_var_img_path)
    vanilla_var_img = Image.open(vanilla_var_img_path)

    original_images.append(original_img)
    var_images.append(var_img)
    mmd_var_images.append(mmd_var_img)
    wasserstein_var_images.append(wasserstein_var_img)
    online_var_images.append(online_var_img)
    ema_var_images.append(ema_var_img)
    vanilla_var_images.append(vanilla_var_img)
    
# Get image size
img_width, img_height = original_images[0].size

# Canvas dimensions
canvas_width = img_width * 8
canvas_height = img_height * 7

# Create canvas
canvas = Image.new('RGB', (canvas_width, canvas_height), 'white')

font_path = os.path.join("C:", "Windows", "Fonts", "times.ttf")
# Font setup
try:
    font = ImageFont.truetype(font_path, size=22)
    print("Font loaded successfully")
except IOError:
    font = ImageFont.load_default()
    print("Font not loaded, using default font")

draw = ImageDraw.Draw(canvas)

# Titles for each row
titles = [
    'Original',
    'VAR',
    'Vanila VAR',
    'EMA VAR',
    'Online VAR',
    'Wasserstein VAR',
    'MMD VAR'
]

# Paste images onto the canvas and add row titles
for row_index, (title, images) in enumerate(zip(titles, [original_images, var_images, vanilla_var_images, ema_var_images, online_var_images, wasserstein_var_images, mmd_var_images])):
    for col_index, img in enumerate(images):
        x_position = col_index * img_width
        y_position = row_index * img_height
        canvas.paste(img, (x_position, y_position))

    # Add the row title at the top-left corner of the first image in the row
    title_position = (5, row_index * img_height + 5)
    draw.text(title_position, title, fill=(255, 255, 255, 255), font=font)

# Save the final image
canvas.save('VAR_Visualization_On_ImageNet.png')
```
