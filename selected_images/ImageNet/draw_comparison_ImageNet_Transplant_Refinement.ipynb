```python
import os
import json
from PIL import Image, ImageDraw, ImageFont
from pprint import pprint
```

```python
manul_picked = [
    "004339.png",
    "013772.png",
    "016954.png",
    "048662.png",
    "047779.png",
    "011531.png",
    "017008.png",
    "016995.png"
]
```

```python
transplant_images = []
refinement_images = []


# Base directory for images
base_dir = 'D:\ResearchProject\MLProjects\VQ-Transplant\\figures\ImageNet'


for key in manul_picked:
    # image path
    transplant_img_path = os.path.join(base_dir, 'MMD_VAR_Transplant', key)
    refinement_img_path = os.path.join(base_dir, 'MMD_VAR', key)

    # Load images
    transplant_img = Image.open(transplant_img_path)
    refinement_img = Image.open(refinement_img_path)

    transplant_images.append(transplant_img)
    refinement_images.append(refinement_img)
    
# Get image size
img_width, img_height = transplant_images[0].size

# Canvas dimensions
canvas_width = img_width * 8
canvas_height = img_height * 2

# Create canvas
canvas = Image.new('RGB', (canvas_width, canvas_height), 'white')

font_path = os.path.join("C:", "Windows", "Fonts", "times.ttf")
# Font setup
try:
    font = ImageFont.truetype(font_path, size=22)
    print("Font loaded successfully")
except IOError:
    font = ImageFont.load_default()
    print("Font not loaded, using default font")

draw = ImageDraw.Draw(canvas)

# Titles for each row
titles = [
    'Transplant',
    'Refinement',
]

# Paste images onto the canvas and add row titles
for row_index, (title, images) in enumerate(zip(titles, [transplant_images, refinement_images])):
    for col_index, img in enumerate(images):
        x_position = col_index * img_width
        y_position = row_index * img_height
        canvas.paste(img, (x_position, y_position))

    # Add the row title at the top-left corner of the first image in the row
    title_position = (5, row_index * img_height + 5)
    draw.text(title_position, title, fill=(255, 255, 255, 255), font=font)

# Save the final image
canvas.save('Transplant_Refinement_Visualization.png')
```
