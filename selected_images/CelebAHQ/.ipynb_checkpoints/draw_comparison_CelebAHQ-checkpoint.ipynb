```python
import os
import json
from PIL import Image, ImageDraw, ImageFont
from pprint import pprint
```

```python
manul_picked = [
    "00080.png",
    "00065.png",
    "00087.png",
    "00015.png",
    "00097.png",
    "00083.png",
    "00020.png",
    "00030.png"
]
```

```python
org_images = []
w_vq_images = []
categorical_images = []


# Base directory for images
base_dir = 'reconstructed_images_new'


for key in manul_picked_new:
    # Original image path
    org_img_path = os.path.join(base_dir, 'raw', key)
    # WassersteinVQ image path
    w_vq_img_path = os.path.join(base_dir, 'rec', key)

    # Load images
    org_img = Image.open(org_img_path)
    w_vq_img = Image.open(w_vq_img_path)

    org_images.append(org_img)
    w_vq_images.append(w_vq_img)
    
for key in manul_picked_new_2:
    categorical_img_path = os.path.join(base_dir, 'categorical_tokenizer', key)
    categorical_img = Image.open(categorical_img_path)
    categorical_images.append(categorical_img)

# Get image size
img_width, img_height = org_images[0].size

# Canvas dimensions
canvas_width = img_width * 8
canvas_height = img_height * 3

# Create canvas
canvas = Image.new('RGB', (canvas_width, canvas_height), 'white')

font_path = os.path.join("C:", "Windows", "Fonts", "times.ttf")
# Font setup
try:
    font = ImageFont.truetype(font_path, size=22)
    print("Font loaded successfully")
except IOError:
    font = ImageFont.load_default()
    print("Font not loaded, using default font")

draw = ImageDraw.Draw(canvas)

# Titles for each row
titles = [
    'Original',
    'VAR-Wasserstein VQ',
    'Categorical Tokenizer'
]

# Paste images onto the canvas and add row titles
for row_index, (title, images) in enumerate(zip(titles, [org_images, w_vq_images, categorical_images])):
    for col_index, img in enumerate(images):
        x_position = col_index * img_width
        y_position = row_index * img_height
        canvas.paste(img, (x_position, y_position))

    # Add the row title at the top-left corner of the first image in the row
    title_position = (5, row_index * img_height + 5)
    draw.text(title_position, title, fill=(255, 255, 255, 255), font=font)

# Save the final image
canvas.save('comparison_new.png')
```
